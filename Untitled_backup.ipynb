```python

def getListDateElements():
    currentMonthPath = '//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[1]/div[1]/div[1]/span[1]'
    currentMonth = driver.find_element("xpath", currentMonthPath)
    currentMonth = currentMonth.get_attribute('innerHTML')
    # print("========================")
    # print("current Month")
    # print(currentMonth)
    # print("========================\n")
    
    weekPath = '//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[2]/div/div[2]/div[2]/*'
    weekElement = driver.find_elements("xpath", weekPath)
    nWeeks = len(weekElement)
    print(f"Number of weeks: ", nWeeks)
    
    listDateElements = []
    for itrWeek in range(1, nWeeks+1):
        
        eachWeekPath = f'//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[2]/div/div[2]/div[2]/div[{itrWeek}]/'
        for itrDay in range(1, 8):
            eachDatePath = f'//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[2]/div/div[2]/div[2]/div[{itrWeek}]/button[{itrDay}]/div[1]/span'
            eachDateElement = driver.find_elements("xpath", eachDatePath)
            listDateElements.append(eachDateElement[0])
    
    listDateElements = [ (eachDate.text, eachDate) for eachDate in listDateElements if len(eachDate.text) > 0 ]
    # print(listDateElements)
    return listDateElements
```

```python
def clickChange(monthElement, listDateElements):
    # changeElement
    changePath = '//*[@id="air-redemption-ui-app"]/div/div[2]/div/div/div[1]/div[1]/div/div/button'
    changeElement = driver.find_element("xpath", changePath)
    # Scroll to the top
    reClickMonth = False
    while(True):
        try:
            driver.find_element(By.TAG_NAME, "body").send_keys(Keys.CONTROL + Keys.HOME)
            time.sleep(2)
            changeElement.click()
            
            break
        except:
            print("Can't click change try again")
            reClickMonth = True
    time.sleep(10)

    if reClickMonth:
        print("Need to get the Month again ... ")
        monthElement.click()
        sleep(5)
        listDateElements = getListDateElements()

    return listDateElements
```

```python
def goThroughDates():
    # Get number of flights 
    
    pathToFlightTable = '//*[@id="flights-firstTable"]/div[3]/*'
    nFlightTable = len(driver.find_elements("xpath", pathToFlightTable))
    print(f"Number of flight tables found: {nFlightTable}")
    
    lsRoutesOps = []
    for itr in range(1, nFlightTable + 1):
        
        # Initial route
        dict_route = {'route':itr}
        
        tableOpenPath = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div/div[2]/div/button/div'
    
        isDirect = True
        # Open the table
        eachTableUnfoldElement = driver.find_elements("xpath", tableOpenPath)
        if len(eachTableUnfoldElement) == 0:
            print(f"Round {itr} Direct Flight")
        else:
            print(f"Round {itr} NOT Direct Flight")
            isDirect = False
            
            # Open Table
            eachTableUnfoldElement[0].click()
            
        dict_route.update({'isDirect':isDirect})
        time.sleep(2)
    
        # Get the departure airport and arrival airpot
        pathToFlights = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/*'
        nFlightsInTrip = len(driver.find_elements("xpath", pathToFlights))
        
        if isDirect:
            nFlightsInTrip = nFlightsInTrip
        else:
            nFlightsInTrip = nFlightsInTrip - 2
        print("    Number of flights in this route: ", nFlightsInTrip)
        
        if isDirect:
            rngSt = 1
            engEd = 1 + nFlightsInTrip
        else:
            rngSt = 2
            engEd = 2 + nFlightsInTrip
            
        for itrFlight in range(rngSt, engEd):
            # print("rngSt, engEd", rngSt, engEd)
            
            if isDirect:
                padd = ''
                dict_route.update({'itrFlight':itrFlight})
                print("    Flight : ", itrFlight)
            else:
                dict_route.update({'itrFlight':itrFlight - 1})
                print("    Flight : ", itrFlight - 1 )
                padd = 'div/'
            # Get Flight departure and arrival airports
            deptAptPath = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div[{itrFlight}]/div[1]/div[1]/div/{padd}div[2]/div[1]/div[2]/div[1]'
            deptAptElemt = driver.find_elements("xpath", deptAptPath)
            deptApt = deptAptElemt[0].get_attribute('innerHTML')
                            
            arrAptPath =  f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div[{itrFlight}]/div[1]/div[1]/div/{padd}div[2]/div[1]/div[2]/div[3]'
            arrAptElemt = driver.find_elements("xpath", arrAptPath)
            arrApt = arrAptElemt[0].get_attribute('innerHTML')
            
            # Get Flight departure and arrival time
            deptAptTimePath = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div[{itrFlight}]/div[1]/div[1]/div/{padd}div[2]/div[1]/div[1]/div[1]'
            deptAptTimeElemt = driver.find_elements("xpath", deptAptTimePath)
            deptAptTime = deptAptTimeElemt[0].get_attribute('innerHTML')
    
            arrAptTimePath = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div[{itrFlight}]/div[1]/div[1]/div/{padd}div[2]/div[1]/div[1]/div[3]'
            arrAptTimeElemt = driver.find_elements("xpath", arrAptTimePath)
            arrAptTime = arrAptTimeElemt[0].get_attribute('innerHTML')
            
            print(f"        deptApt: {deptApt}, deptAptTime: {deptAptTime}, arrApt: {arrApt}, arrAptTime: {arrAptTime}" )
            dict_route.update({'deptApt':deptApt, 'deptAptTime': deptAptTime, 'arrApt':arrApt, 'arrAptTime':arrAptTime})
                
            # Get flight info
            if isDirect:
                padd = ''
            else:
                padd = 'div/'
            
            flightNoPath = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div[{itrFlight}]/div[1]/div[1]/div/{padd}div[2]/div[2]/{padd}{padd}div/p[1]'
            flightNoElemt = driver.find_elements("xpath", flightNoPath)
            flightNo = flightNoElemt[0].get_attribute('innerHTML')
            
    
            flightSpanPath = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div[{itrFlight}]/div[1]/div[1]/div/{padd}div[2]/div[2]/{padd}{padd}div/span'
            flightSpanElemt = driver.find_elements("xpath", flightSpanPath)
            flightSpan = flightSpanElemt[0].get_attribute('innerHTML')
            
    
            operateByPath = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div[{itrFlight}]/div[1]/div[1]/div/{padd}div[2]/div[2]/{padd}{padd}div/p[2]'
            operateByElemt = driver.find_elements("xpath", operateByPath)
            operateBy = operateByElemt[0].get_attribute('innerHTML').replace('Operated by ','')
            
            
            print(f"        flightNo: {flightNo}, flightSpan: {flightSpan}, operateBy: {operateBy}" )
            print("\n")
            dict_route.update({'flightNo':flightNo, 'flightSpan': flightSpan, 'operateBy':operateBy})
            if isDirect:
                padd = ''
            else:
                padd = 'div[1]/'
            # Get miles option
            clsOpsPath = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div[{itrFlight}]/{padd }div[2]/div/div/*'
            clsOpsElemt = driver.find_elements("xpath", clsOpsPath)
            nClsOps = len(clsOpsElemt)
            
            print(f"        Number Class Options: {nClsOps}" )
            
            for itrOption in range(1, nClsOps + 1):
                
                opsPath = f'//*[@id="flights-firstTable"]/div[3]/div[{itr}]/div[{itrFlight}]/{padd }div[2]/div/div/div[{itrOption}]'
    
                # Get number of mile Info
                milePath = opsPath + '/div/button/div/div/p'
                mileElemt = driver.find_elements("xpath", milePath)
                nMile = len(mileElemt)
    
                if nMile > 0:
                    mileOpsSeats = []
                    for itrMiles in range(1, nMile + 1):
                        milePath = opsPath + f'/div/button/div/div/p[{itrMiles}]'
                        mileElemt = driver.find_elements("xpath", milePath)
                        mileOp = mileElemt[0].get_attribute('innerHTML')
                        mileOpsSeats.append(mileOp)
                else:
                    mileOpsSeats = []
                    mileOpsSeats.append(None)
                print(f"        itrOption: {itrOption} number of miles info {nMile}: {str(tuple(mileOpsSeats))}")
                
                dict_route_ops = copy.deepcopy(dict_route)
                if nMile == 2:
                    dict_route_ops.update({'itrClass':itrOption, 'ClassMiles':mileOpsSeats[0], 'ClassInfo':mileOpsSeats[1]})
                    print(f"        Class Options: {itrOption}, miles: {mileOpsSeats[0]}, milesInfo: {mileOpsSeats[1]}" )
                else:
                    dict_route_ops.update({'itrClass':itrOption, 'ClassMiles':mileOpsSeats[0]})
                    print(f"        Class Options: {itrOption}, miles: {mileOpsSeats[0]}" )
                lsRoutesOps.append(dict_route_ops)
    
            print("\n")   
            #print(f"Trip {itr}, fligth {itrFlight - 1}: from {deptApt} at {deptAptTime} TO {arrApt} at {arrAptTime}")
            #print(" ")
            #break                                    
        print("\n\n")
        # Close the table
        if len(eachTableUnfoldElement) != 0:
            eachTableUnfoldElement = driver.find_element("xpath", tableOpenPath)
            eachTableUnfoldElement.click()
        time.sleep(2)
    
        #break
    
    dfRoutesOps = pd.DataFrame(lsRoutesOps)
    display(dfRoutesOps)
    return dfRoutesOps
```

```python

for itrMonth in range(itrStartMonth, itrEndMonth+1):
    
    month, monthElement = listMonthElement[itrMonth]

    print(f"Process {month} .................. ")
    monthElement.click()
    sleep(10)

    listDateElements = getListDateElements()

    for itrDate in range(0, len(listDateElements)):

        
        date, dateElement = listDateElements[itrDate]

        print(f"    Process date {date}  ")

        while(True):
            try:
                dateElement.click()
                time.sleep(15)
                break
            except:
                print("Date Click Fail. Try again ...")
                monthElement.click()
                sleep(10)
                listDateElements = getListDateElements()
                date, dateElement = listDateElements[itrDate]

        with io.capture_output() as captured:
            dfRoutesOps = goThroughDates()
            
        dfRoutesOps['Month'] = month
        dfRoutesOps['month'] = dictMonth[month]
        dfRoutesOps['Date'] = date
        print(f"        number rows: {dfRoutesOps.shape[0]}")
        tolRoutes.append(dfRoutesOps)

        clickChange(monthElement, listDateElements)
        
        pd.concat(tolRoutes, ignore_index=True).to_csv('tickets.csv')


```
