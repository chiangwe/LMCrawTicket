```python

def getListDateElements(month):
    # Check Current Month
    monthItrOffset = 2
    #currentMonthPath = '//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[1]/div[1]/div[1]/span[1]'
    currentMonthPath = f'//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[2]/div/div[{monthItrOffset}]/div[1]/span[1]'
    currentMonth = driver.find_element("xpath", currentMonthPath)
    currentMonth = currentMonth.get_attribute('innerHTML')

    if currentMonth != month:
        monthItrOffset = 3
        currentMonthPath = f'//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[2]/div/div[{monthItrOffset}]/div[1]/span[1]'
        currentMonth = driver.find_element("xpath", currentMonthPath)
        currentMonth = currentMonth.get_attribute('innerHTML')
        if currentMonth!=month:
            raise Exception("Current Month Is not Correct.")

    print("monthItrOffset: ", monthItrOffset)

    # print("========================")
    # print("current Month")
    # print(currentMonth)
    # print("========================\n")
    
    weekPath = f'//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[2]/div/div[{monthItrOffset}]/div[2]/*'
    weekElement = driver.find_elements("xpath", weekPath)
    nWeeks = len(weekElement)
    print(f"Number of weeks: ", nWeeks)
    
    listDateElements = []
    for itrWeek in range(1, nWeeks+1):
        
        eachWeekPath = f'//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[2]/div/div[{monthItrOffset}]/div[2]/div[{itrWeek}]/'
        for itrDay in range(1, 8):
            eachDatePath = f'//*[@id="RedemptionActiveWrapper"]/div[2]/div/div/div[3]/div[2]/div[2]/div/div[{monthItrOffset}]/div[2]/div[{itrWeek}]/button[{itrDay}]/div[1]/span'
            #print(eachDatePath)
            eachDateElement = driver.find_elements("xpath", eachDatePath)
            listDateElements.append(eachDateElement[0])
    
    listDateElements = [ (eachDate.text, eachDate) for eachDate in listDateElements if len(eachDate.text) > 0 ]
    # print(listDateElements)
    return listDateElements
```

```python
def clickChange(month, monthElement, listDateElements):
    # changeElement
    changePath = '//*[@id="air-redemption-ui-app"]/div/div[2]/div/div/div[1]/div[1]/div/div/button'
    changeElement = driver.find_element("xpath", changePath)

    # Scroll to the top
    reClickMonth = False
    while(True):

        try:
            driver.execute_script("window.scrollTo(0, document.body.scrollTop);")
            time.sleep(2)
            changeElement.click()
            break
        except Exception as err:
            print("Can't click change try again", err)
            reClickMonth = True
    time.sleep(10)

    if reClickMonth:
        print("Need to get the Month again ... ")
        monthElement.click()
        sleep(5)
        listDateElements = getListDateElements(month)

    return listDateElements
```

```python

# Set up parameters

# Set up airport
CrawlDate = '03_06_2024'
fromAir = 'SFO'
toAir = 'NRT'

startMonth = 'May'
# startMonth = 'Sep'
endMonth = 'Sep'

# Create output folder
pathOut = './output/' + CrawlDate + '/'
os.makedirs(pathOut, exist_ok=True)


# We change account after each month
itrStartMonth = dictMonth[startMonth]
itrEndMonth = dictMonth[endMonth]

# Get account 
nAccount = len(lsAccount)
itrAccount = 15
accountName = lsAccount[itrAccount]
account = dictAccount[accountName]['userName']
cred = dictAccount[accountName]['credentials']

for itrMonth in range(itrStartMonth-1, itrEndMonth+1):
    
    month = listMonth[itrMonth]
    print(f"Process {month} .................. ")
    pathCSV = pathOut + f'ticket_{fromAir}_{toAir}_{month}.csv'
    
    driver = startCrome()
    driver = getUrl(driver)
    driver = changeLang(driver)
    driver = logIn(driver, account, cred)
    driver = remindMeLater(driver)
    driver = checkAlias(driver)
    driver = changeConfig(driver)
    driver = setUpAirport(driver, fromAir, toAir)

    # Get month
    listMonthElement, driver = getMonths(driver)
    dictMonthElement = dict(listMonthElement)
    monthElement = dictMonthElement[month]
    
    # Click on month
    monthElement.click()
    sleep(10)
    
    listDateElements = getListDateElements(month)
    print([each[0] for each in listDateElements])
    
    monthCSV = []
    for itrDate in range(0, len(listDateElements)):

        process = True
        while(process):
            
            date, dateElement = listDateElements[itrDate]

            print(f"    Process date {date}  out of {len(listDateElements)}")
        
            try:
                dateElement.click()
                
                CheckDepartureFlighPath = '//*[@id="RedemptionActiveWrapper"]/div[2]'
                WebDriverWait(driver, 30).until(EC.presence_of_element_located(("xpath", CheckDepartureFlighPath)))
                time.sleep(15)

                with io.capture_output() as captured:
                #while(True):
                    try:
                        #with io.capture_output() as captured:
                        dfRoutesOps = goThroughDates()
                        process = False
                        break
                    except Exception as error:
                        print("Flight Table Click Fail. Try again ...")
                        print(error)
                        listDateElements = clickChange(month, monthElement, listDateElements)
                        raise Exception(error)
                        break
                        # date, dateElement = listDateElements[itrDate]
            except Exception as error:
                print("Date Click Fail. Try again ...")
                print(error)
                monthElement.click()
                sleep(10)
                listDateElements = getListDateElements()
                # date, dateElement = listDateElements[itrDate]

     
        dfRoutesOps['Month'] = month
        dfRoutesOps['month'] = dictMonth[month]
        dfRoutesOps['Date'] = date
        print(f"        number rows: {dfRoutesOps.shape[0]}")
        tolRoutes.append(dfRoutesOps)

        listDateElements = clickChange(month, monthElement, listDateElements)

        # Chek the first class and send 
        if dfRoutesOps.shape[0] > 0:
            getDirect = dfRoutesOps[dfRoutesOps['isDirect']]
            getDirect = getDirect[getDirect['itrClass'] == 2]
            getDirect = getDirect[getDirect['ClassMiles'].notnull()]
            getDirect = [ json.dumps(each) for each in getDirect.to_dict('records')]
    
            for eachJson in getDirect:
                payload['body'] = eachJson
                response = requests.post(url, json=payload, headers=headers)
    
                print(payload)
                print(response.json())
                print(payload)
            
            
        monthCSV.append(dfRoutesOps)
        pd.concat(tolRoutes, ignore_index=True).to_csv('tickets_temp.csv', index=False)

    # Save Month CSV
    monthCSV = pd.concat(monthCSV, ignore_index=True)
    monthCSV.to_csv(pathCSV, index=False)
    
    if itrAccount == (nAccount - 1):
        itrAccount = 1
    else:
        itrAccount = itrAccount + 1
    driver.close()

```
